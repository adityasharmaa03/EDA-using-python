```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Loading Dataset

```python
data = pd.read_csv('hotel_booking.csv')
```

### Performing EDA and Data Cleaning

```python
data.head()
```

```python
data.shape
```

```python
# removing unnecessary data such as name of customer, email and other personal data
data = data.drop(['name', 'email', 'phone-number', 'credit_card'], axis=1)
```

```python
data.shape
```

```python
# getting the names of columns in the dataset
data.columns
```

```python
# checking the datatype of columns
data.info()
```

```python
# converting reservation_status_date from object type to datetime
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'])
```

```python
data.info()
```

```python
# checking the categorical data i.e. object datatype
data.describe(include= 'object')
```

```python
# checking each category present in this object datatype
for col in data.describe(include= 'object').columns:
    print(col)
    print(data[col].unique())
    print('*' * 50)
```

```python
# checking the null values
data.isnull().sum()
```

```python
# drop the column 'agent' and 'company'
data.drop(['agent', 'company'], axis=1, inplace=True)

# removing the missing values from data
data.dropna(inplace=True)
```

```python
# again checking the missing values
data.isnull().sum()
```

```python
# checking statistic summary of the data
data.describe()
```

```python
# some outliers are present in the data such as in adr column 5400 is an outlier
data['adr'].plot(kind= 'box')
```

```python
# removing the outlier from adr
data = data[data['adr']<5000]
```

```python
data.describe()
```

## Data Analysis and Visualization

```python
# checking cancel percentage
cancelled_percent = data['is_canceled'].value_counts(normalize=True)
print(cancelled_percent*100)

# visualize this data
plt.figure(figsize = (5, 4))
plt.title('Reservation Status')
plt.bar(['Not Cancelled', 'Cancelled'], data['is_canceled'].value_counts(), edgecolor='k', width = 0.7)
plt.show()
```

```python
# cancelation rate depending on hotels
plt.figure(figsize = (8, 4))
ax1 = sns.countplot(x = 'hotel', hue='is_canceled', data = data, palette = 'Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
# ax1.legend(bbox_to_anchor(1,1))
plt.title('Reservation Cancelation in Hotels')
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
# checking percent of each hotel
resort_hotel = data[data['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python

city_hotel = data[data['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# checking the affect of price on booking cancellation
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# visualizing it
plt.figure(figsize = (20, 8))
plt.title('Average daily rates of hotels', fontsize = 18)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize = 16)
plt.show()
```

```python
# lets check the cancelation and reservation month-wise
data['month'] = data['reservation_status_date'].dt.month
plt.figure(figsize = (20, 8))
ax1 = sns.countplot(x = 'month', hue='is_canceled', data = data, palette = 'bright')
legend_labels, _ = ax1.get_legend_handles_labels()
plt.title('Reservation and Cancelation month-wise')
plt.xlabel('Month')
plt.ylabel('Number of Reservation')
plt.legend(['Not-canceled', 'Canceled'])
plt.show()
```

```python
# avg daily rates for each moth
df = data
plt.figure(figsize = (20, 8))
plt.title('ADR for each Month')
sns.barplot(x = 'month', y = 'adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

## This proves that the cancellation is due to the higher rates. As the rates are higher in 1 month and cancellation also higher and in month 8 rates
## are low so cancelation is also low.

### Top 10 countries with highest booking cancellation

```python
cancel_data = data[data['is_canceled'] == 1]
top_ten_countries = cancel_data['country'].value_counts()[:10]
plt.figure(figsize= (8, 8))
plt.title('Top 10 Countries with highest Cancellation')
plt.pie(top_ten_countries, autopct='%.2f', labels = top_ten_countries.index)
plt.show()
```
